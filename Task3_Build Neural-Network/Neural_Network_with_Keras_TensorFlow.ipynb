```python
# ========================================
# FIRST: Run this installation cell separately
# ========================================
# !pip install tensorflow matplotlib

# ========================================
# THEN: Run this complete solution in one cell
# ========================================

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

print("üöÄ Starting MNIST Neural Network Training...")
print("TensorFlow version:", tf.__version__)

# ========================================
# STEP 1: LOAD AND PREPROCESS DATA
# ========================================
print("\n=== STEP 1: Loading MNIST Dataset ===")
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")

# Show sample images
plt.figure(figsize=(12, 4))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f'Label: {y_train[i]}')
    plt.axis('off')
plt.suptitle('Sample MNIST Digits', fontsize=16)
plt.tight_layout()
plt.show()

# Normalize pixel values to 0-1 range
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Convert labels to categorical one-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print(f"‚úÖ Data preprocessed: {X_train.shape} training samples, {X_test.shape} test samples")

# ========================================
# STEP 2: BUILD NEURAL NETWORK MODEL
# ========================================
print("\n=== STEP 2: Building Neural Network Model ===")

model = Sequential([
    Flatten(input_shape=(28, 28)),           # Flatten 28x28 images to 784 features
    Dense(128, activation='relu'),            # Hidden layer with 128 neurons + ReLU
    Dropout(0.2),                            # Dropout layer (20%) to prevent overfitting
    Dense(10, activation='softmax')           # Output layer: 10 neurons for digits 0-9
])

# Display model architecture
print("\nüìã Model Architecture:")
model.summary()

# ========================================
# STEP 3: COMPILE MODEL
# ========================================
print("\n=== STEP 3: Compiling Model ===")

model.compile(
    optimizer='adam',                         # Adam optimizer
    loss='categorical_crossentropy',          # Loss function for multi-class classification
    metrics=['accuracy']                      # Track accuracy during training
)

print("‚úÖ Model compiled successfully!")

# ========================================
# STEP 4: TRAIN MODEL
# ========================================
print("\n=== STEP 4: Training Model ===")
print("Training for 10 epochs... (this may take 2-3 minutes)")

history = model.fit(
    X_train, y_train,
    epochs=10,                               # Train for 10 epochs
    batch_size=32,                           # Process 32 samples at a time
    validation_data=(X_test, y_test),        # Use test data for validation
    verbose=1                                # Show progress
)

print("‚úÖ Training completed!")

# ========================================
# STEP 5: EVALUATE MODEL
# ========================================
print("\n=== STEP 5: Model Evaluation ===")

test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"üìä Final Test Results:")
print(f"   ‚Ä¢ Test Loss: {test_loss:.4f}")
print(f"   ‚Ä¢ Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

# ========================================
# STEP 6: VISUALIZE TRAINING HISTORY
# ========================================
print("\n=== STEP 6: Plotting Training History ===")

# Create comprehensive training plots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Training & Validation Accuracy
ax1.plot(history.history['accuracy'], 'b-o', label='Training Accuracy', linewidth=2, markersize=6)
ax1.plot(history.history['val_accuracy'], 'r-s', label='Validation Accuracy', linewidth=2, markersize=6)
ax1.set_title('Model Accuracy Over Time', fontsize=14, fontweight='bold')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim(0, 1)

# Plot 2: Training & Validation Loss
ax2.plot(history.history['loss'], 'b-o', label='Training Loss', linewidth=2, markersize=6)
ax2.plot(history.history['val_loss'], 'r-s', label='Validation Loss', linewidth=2, markersize=6)
ax2.set_title('Model Loss Over Time', fontsize=14, fontweight='bold')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Accuracy Only (Cleaner view)
ax3.plot(range(1, 11), history.history['accuracy'], 'g-o', label='Training', linewidth=3, markersize=8)
ax3.plot(range(1, 11), history.history['val_accuracy'], 'orange', marker='s', label='Validation', linewidth=3, markersize=8)
ax3.set_title('Accuracy Comparison', fontsize=14, fontweight='bold')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Accuracy')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: Loss Only (Cleaner view)
ax4.plot(range(1, 11), history.history['loss'], 'purple', marker='o', label='Training', linewidth=3, markersize=8)
ax4.plot(range(1, 11), history.history['val_loss'], 'brown', marker='s', label='Validation', linewidth=3, markersize=8)
ax4.set_title('Loss Comparison', fontsize=14, fontweight='bold')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Loss')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ========================================
# STEP 7: TEST PREDICTIONS
# ========================================
print("\n=== STEP 7: Testing Model Predictions ===")

# Make predictions on test set
predictions = model.predict(X_test[:12], verbose=0)
predicted_classes = np.argmax(predictions, axis=1)
actual_classes = np.argmax(y_test[:12], axis=1)

# Display sample predictions with images
plt.figure(figsize=(15, 8))
for i in range(12):
    plt.subplot(3, 4, i+1)
    plt.imshow(X_test[i], cmap='gray')

    # Color code: Green for correct, Red for incorrect
    is_correct = predicted_classes[i] == actual_classes[i]
    color = 'green' if is_correct else 'red'
    confidence = np.max(predictions[i]) * 100

    plt.title(f'Pred: {predicted_classes[i]} | True: {actual_classes[i]}\n'
              f'Confidence: {confidence:.1f}%',
              color=color, fontweight='bold', fontsize=10)
    plt.axis('off')

plt.suptitle('Sample Test Predictions\n(Green = Correct, Red = Incorrect)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Print detailed predictions
print("\nDetailed Prediction Results:")
correct_predictions = 0
for i in range(12):
    confidence = np.max(predictions[i]) * 100
    is_correct = predicted_classes[i] == actual_classes[i]
    if is_correct:
        correct_predictions += 1

    status = "‚úÖ CORRECT" if is_correct else "‚ùå WRONG"
    print(f"Image {i+1:2d}: Predicted = {predicted_classes[i]} | Actual = {actual_classes[i]} | "
          f"Confidence = {confidence:5.1f}% | {status}")

sample_accuracy = (correct_predictions / 12) * 100
print(f"\nSample Accuracy: {correct_predictions}/12 ({sample_accuracy:.1f}%)")

# ========================================
# FINAL SUMMARY
# ========================================
print(f"\nüéâ ============== TRAINING COMPLETE ==============")
print(f"üìä FINAL RESULTS:")
print(f"   ‚Ä¢ Model Architecture: Flatten ‚Üí Dense(128) ‚Üí Dropout(0.2) ‚Üí Dense(10)")
print(f"   ‚Ä¢ Training Samples: {X_train.shape[0]:,}")
print(f"   ‚Ä¢ Test Samples: {X_test.shape[0]:,}")
print(f"   ‚Ä¢ Training Epochs: 10")
print(f"   ‚Ä¢ Total Parameters: {model.count_params():,}")
print(f"   ‚Ä¢ Final Test Accuracy: {test_accuracy*100:.2f}%")
print(f"   ‚Ä¢ Final Test Loss: {test_loss:.4f}")

# Training progress summary
final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
final_train_loss = history.history['loss'][-1]
final_val_loss = history.history['val_loss'][-1]

print(f"\nüìà TRAINING PROGRESS:")
print(f"   ‚Ä¢ Final Training Accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)")
print(f"   ‚Ä¢ Final Validation Accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)")
print(f"   ‚Ä¢ Final Training Loss: {final_train_loss:.4f}")
print(f"   ‚Ä¢ Final Validation Loss: {final_val_loss:.4f}")

# Check for overfitting
accuracy_gap = final_train_acc - final_val_acc
if accuracy_gap > 0.05:
    print(f"‚ö†Ô∏è  WARNING: Possible overfitting detected (gap: {accuracy_gap:.4f})")
else:
    print("‚úÖ Model shows good generalization!")

print(f"\nüèÜ SUCCESS: Neural network trained successfully with {test_accuracy*100:.2f}% accuracy!")
print("="*60)
```
